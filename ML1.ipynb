# 4. Fill missing values in the age column using mean imputation.

What it does:
df['Age'].fillna(df['Age'].mean()):

.This replaces all missing values in the Age column with the mean (average) age.

.It's a common strategy when you're dealing with missing numeric values.

df.isnull().sum():

.After the replacement, this shows how many missing values are left in each column.

```python
df['Age']=df['Age'].fillna(df['Age'].mean())

df.isnull().sum()
```

# 5. Fill missing values in the fare column using median imputation.

df['Fare'].median() computes the median fare (the middle value, not affected by extreme values).

.fillna(...) replaces any missing NaN values in the Fare column with that median.

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
df.isnull().sum()
```

# 6. Fill missing values in the embarked column using mode imputation.

Replaces missing values in the Embarked column with the most frequent port (mode).

Then shows the total count of missing values in each column.

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df.isnull().sum()
```

# 7. Fill missing values in the cabin column with the constant "Unknown".

Replaces all NaN values in the Cabin column with the constant string 'unknown'.

Then checks how many missing values remain in each column.



```python
df['Cabin'] = df['Cabin'].fillna('unknown')
df.isnull().sum()
```

# 8. Apply forward fill on the age column.


method='ffill' (forward fill) fills missing values by carrying the last valid value downward.

This is another common imputation method, especially for time-series or ordered data.

Then df.isnull().sum() shows how many missing values are left.

```python
df['Age']= df['Age'].fillna(method='ffill')
df.isnull().sum()
```

# 9. Apply backward fill on the age column.

method='bfill' (backward fill) tells pandas:

“If a value is missing, fill it using the next valid value below it.”

Then df.isnull().sum() prints how many missing values remain.

```python
df['Age']= df['Age'].fillna(method='bfill')
df.isnull().sum()
```

# Section 4: Feature Scaling
12. Apply Min-Max Scaling to the age column and create a new column age_minmax.




```python
df['age_minmax']=(df['Age']-df['Age'].min())/(df['Age'].max()-df['Age'].min())
df.head()
```

13. Apply Z-score standardization to the fare column and create a column fare_z.


```python
df['fare_z']=(df['Fare']-df['Fare'].mean())/df['Fare'].std()
df.head()
```

14. Apply Robust Scaling to the fare column and create fare_robust.


```python
df['fare_robust']=(df['Fare']-df['Fare'].median())/(df['Fare'].quantile(0.75)-df['Fare'].quantile(0.25))
df.head()
```

15. Apply Max-Abs Scaling to the fare column and create fare_maxabs

```python
df['fare_maxabs']=(df['Fare']-df['Fare'].min())/(df['Fare'].abs().max())
df.head(25)
```

# Section 5: Outlier Detection and Removal


16. Detect and remove outliers in the age column using the IQR method.


```python
# Calculate the IQR for the 'Age' column
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify and remove outliers
df_no_outliers = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)].copy()

print(f"Shape of original DataFrame: {df.shape}")
print(f"Shape of DataFrame after removing outliers: {df_no_outliers.shape}")

display(df_no_outliers.head())
```

17. Detect and remove outliers in the age column using the Z-score method.

```python
from scipy.stats import zscore

# Calculate the Z-scores for the 'Age' column
z_scores = np.abs(zscore(df['Age']))

# Define a threshold for outlier detection (commonly 3)
threshold = 3

# Identify and remove outliers
df_no_outliers_zscore = df[z_scores < threshold].copy()

print(f"Shape of original DataFrame: {df.shape}")
print(f"Shape of DataFrame after removing outliers (Z-score): {df_no_outliers_zscore.shape}")

display(df_no_outliers_zscore.head())
```
